## Descargar datos y realizar una llamada a la API

De la documentacion oficial podemos sacar el siguiente código para realizar el llamado a la API.

```python
# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
CSV_URL = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY_EXTENDED&symbol=XOM&interval=1min&slice=year2month1&apikey=KEY'

with requests.Session() as s:
    download = s.get(CSV_URL)
    decoded_content = download.content.decode('utf-8')
    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
    for row in my_list:
        print(row)
```

El mismo hace referencia a la función de **Intraday (Extended History)**.
Esta API devuelve series de tiempo históricas intradiarias para los últimos 2 años, cubriendo más de 2 millones de puntos de datos por ticker.

Por practicidad **utilicé el slicer de “year2month1” el mismo incluye 1 año y 1 mes completos con un intervalo de actualización de 1 min** (la API permite 1min, 5min, 15min, 30min, 60min)

```python
df = pd.DataFrame(my_list)
df.info()
```

```python
df.head()
```

```python
headers = df.iloc[0]
df2  = pd.DataFrame(df.values[1:], columns=headers)
df2.info()
```

Como se puede ver, este dataframe cumple con el número de filas pero no el de columnas.

## Lista de ETFs

```python
#!/usr/bin/env python

try:
    # For Python 3.0 and later
    from urllib.request import urlopen
except ImportError:
    # Fall back to Python 2's urllib2
    from urllib2 import urlopen

import certifi
import json

def get_jsonparsed_data(url):
    """
    Receive the content of ``url``, parse it as JSON and return the object.

    Parameters
    ----------
    url : str

    Returns
    -------
    dict
    """
    response = urlopen(url, cafile=certifi.where())
    data = response.read().decode("utf-8")
    return json.loads(data)

url = ("https://financialmodelingprep.com/api/v3/etf/list?apikey=KEY")

data_etf = get_jsonparsed_data(url)
print(get_jsonparsed_data(url))
```
